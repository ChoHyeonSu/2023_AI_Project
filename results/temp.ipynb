```python
import pandas as pd

origin = pd.read_csv('../data/step2/song_prob.csv')
base1 = pd.read_csv('baseline_2304.csv')
base2 = pd.read_csv('baseline2_2304.csv')
ours = pd.read_csv('our_2304.csv')
```

```python
# origin에서 필요한 칼럼 선택
origin_selected = origin[['artist', 'song_name', 'Emotion', 'probability']].rename(columns={'Emotion': 'golden_label', 'probability': 'golden_dist'})


# base1, base2, ours에서 'probability' 칼럼 선택 및 칼럼명 변경
base1_selected = base1[['probability']].rename(columns={'probability': 'base1_dist'})
base2_selected = base2[['probability']].rename(columns={'probability': 'base2_dist'})
ours_selected = ours[['probability']].rename(columns={'probability': 'ours_dist'})
```

```python
combined_df = pd.concat([origin_selected, base1_selected, base2_selected, ours_selected], axis=1)
combined_df.to_csv('combined_dist.csv',index=False)
```
