```python
import pandas as pd

original = pd.read_json('song_meta.json')

df = original[['id','album_name','song_name']]

df.to_csv('song_id_albumname_name.csv', index= False)
```

```python
df['issue_date'] = original['issue_date'].astype(str)
df['year'] = df['issue_date'].str[:4].astype(int)

year_filtered = df[df['year']>2017]
```
